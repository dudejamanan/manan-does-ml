```python
#multiple plotting
year = range(2000,2012)
apples = [0.895,0.91,0.926,0.929,0.931,0.934,0.936,0.937,0.94,0.90,0.86,0.83]
oranges = [0.962,0.941,0.930,0.923,0.918,0.908,0.904,0.86,0.80,0.77,0.70,0.65]
```

```python
plt.plot(year,apples)
plt.plot(year,oranges)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
```

```python
#adding legend and title
plt.plot(year,apples)
plt.plot(year,oranges)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.title('Crop Yields in Kanto')
plt.legend(['Apples','Oranges'])
```

```python
#Line Markers
plt.plot(year,apples,marker='o')
plt.plot(year,oranges,marker='x')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
```

```python
#styling lines and markers

#color or c - sets the color of the line
#linestyle or ls - choose between a solid or dashed line
#linewidth or lw - sets the width of a line
#markersize or ms - sets the size of markers
#markeredgecolor or mec - sets the edge color for markers
#markeredgewidth or mew - sets the edge width for markers
#markerfacecolor or mfc - sets the fill color for markers
#alpha - opacity of the plot
```

```python
plt.plot(year,apples,marker='s',c='b',ls='-',lw=2,ms=8,mew=2,mec='y')
plt.plot(year,oranges,marker='o',c='r',ls='--',lw=3,ms=10,mew=3,alpha=.5)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title('Crop Yields in Kanto')
plt.legend(['Apples','Oranges'])
```

```python
#fmt - fmt argument provides a shorthand for specifying the line style,marker and line color.
#(linestyle marker color)
#just a shortcut to fasten the process
plt.plot(year,apples,'s-b') #(square,-,blue)
plt.plot(year,oranges,'o--r') #(circle,--,red)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title('Crop Yields in Kanto')
plt.legend(['Apples','Oranges'])
```

```python
plt.plot(year,oranges,'or') #if we dont add the linestyle it just marks the points

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title('Crop Yields in Kanto')
```

```python
#changing the figure size
plt.figure(figsize=(4,5)) #(9,5) is default
plt.plot(year,oranges,'or')
plt.title('Crop Yields in Kanto')
```
