```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flowers_df = sns.load_dataset('iris') #seaborn can also load data
```

```python
flowers_df.head()
```

```python
flowers_df.shape
```

```python
flowers_df.species.unique()
```

```python
#line chart
plt.plot(flowers_df.sepal_length,flowers_df.sepal_width);
#this does not give any idea about the pattern or something , as we have 150 value which are increasing and decreasing
```

```python
#Scatter plot
sns.scatterplot(x=flowers_df.sepal_length,y=flowers_df.sepal_width);
#here instead of line you will have points marked (150 points)
```

```python
#Adding hue
sns.scatterplot(x=flowers_df.sepal_length,y=flowers_df.sepal_width,hue=flowers_df.species,s=100)
#adding hue so that we can distinguish and find a pattern between different species
```

```python
#Adding title and increase the size
plt.figure(figsize=(12,6));
plt.title("Sepal Dimensions");
sns.scatterplot(x=flowers_df.sepal_length,y=flowers_df.sepal_width,hue = flowers_df.species,s=100);


```
